#plot distributions

```python
k=1
plt.figure(figsize=(40, 40))
for col in df:
  if col=="Attrition":
    continue
  yes = df[df['Attrition'] == 'Yes'][col]
  no = df[df['Attrition'] == 'No'][col]
  plt.subplot(6, 6, k)
  plt.hist(yes, bins=25, alpha=0.5, label='yes', color='b')
  plt.hist(no, bins=25, alpha=0.5, label='no', color='r')
  plt.legend(loc='upper right')
  plt.title(col)
  k+=1
```

```python
# Import necessary libraries
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display

# Create a function to make predictions (you'll need to replace this with your actual model)
def predict_attrition(Age, BusinessTravel, DailyRate, Department, DistanceFromHome,
                     Education, EnvironmentSatisfaction, JobInvolvement, JobLevel,
                     JobSatisfaction, MonthlyIncome, OverTime, WorkLifeBalance):
    """
    This is a placeholder function - you'll need to replace it with your actual model prediction
    """
    # This is just a dummy prediction - replace with your model's prediction logic
    risk_factors = 0

    # Simple heuristic (replace with actual model)
    if OverTime == 'Yes':
        risk_factors += 0.3
    if EnvironmentSatisfaction < 3:
        risk_factors += 0.2
    if JobSatisfaction < 3:
        risk_factors += 0.2
    if WorkLifeBalance < 3:
        risk_factors += 0.2
    if MonthlyIncome < 5000:
        risk_factors += 0.1

    attrition_prob = min(0.95, max(0.05, risk_factors))  # Cap between 5-95%

    if attrition_prob > 0.5:
        return f"High risk of attrition ({attrition_prob:.0%} probability)"
    else:
        return f"Low risk of attrition ({attrition_prob:.0%} probability)"

# Create interactive widgets
style = {'description_width': '200px'}
layout = widgets.Layout(width='400px')

age_slider = widgets.IntSlider(min=18, max=60, step=1, value=35, description='Age:', style=style, layout=layout)
business_travel = widgets.Dropdown(options=['Travel_Rarely', 'Travel_Frequently', 'Non-Travel'],
                                  value='Travel_Rarely', description='Business Travel:', style=style, layout=layout)
daily_rate = widgets.IntSlider(min=100, max=1500, step=100, value=800, description='Daily Rate:', style=style, layout=layout)
department = widgets.Dropdown(options=['Sales', 'Research & Development', 'Human Resources'],
                             value='Research & Development', description='Department:', style=style, layout=layout)
distance_home = widgets.IntSlider(min=1, max=30, step=1, value=10, description='Distance From Home:', style=style, layout=layout)
education = widgets.Dropdown(options=[(str(i), i) for i in range(1, 6)], value=2, description='Education Level:', style=style, layout=layout)
env_satisfaction = widgets.Dropdown(options=[(str(i), i) for i in range(1, 5)], value=3, description='Environment Satisfaction:', style=style, layout=layout)
job_involvement = widgets.Dropdown(options=[(str(i), i) for i in range(1, 5)], value=3, description='Job Involvement:', style=style, layout=layout)
job_level = widgets.Dropdown(options=[(str(i), i) for i in range(1, 6)], value=2, description='Job Level:', style=style, layout=layout)
job_satisfaction = widgets.Dropdown(options=[(str(i), i) for i in range(1, 5)], value=3, description='Job Satisfaction:', style=style, layout=layout)
monthly_income = widgets.IntSlider(min=1000, max=20000, step=500, value=5000, description='Monthly Income:', style=style, layout=layout)
overtime = widgets.Dropdown(options=['Yes', 'No'], value='No', description='Over Time:', style=style, layout=layout)
work_life_balance = widgets.Dropdown(options=[(str(i), i) for i in range(1, 5)], value=3, description='Work Life Balance:', style=style, layout=layout)

# Create the interactive interface
def interactive_predict(Age, BusinessTravel, DailyRate, Department, DistanceFromHome,
                       Education, EnvironmentSatisfaction, JobInvolvement, JobLevel,
                       JobSatisfaction, MonthlyIncome, OverTime, WorkLifeBalance):
    prediction = predict_attrition(Age, BusinessTravel, DailyRate, Department, DistanceFromHome,
                                 Education, EnvironmentSatisfaction, JobInvolvement, JobLevel,
                                 JobSatisfaction, MonthlyIncome, OverTime, WorkLifeBalance)
    print("\nPrediction Result:")
    print("==================")
    print(prediction)

# Display the interface
print("Employee Attrition Risk Predictor")
print("=================================")
print("Please provide the employee's details below:")

interact(interactive_predict,
         Age=age_slider,
         BusinessTravel=business_travel,
         DailyRate=daily_rate,
         Department=department,
         DistanceFromHome=distance_home,
         Education=education,
         EnvironmentSatisfaction=env_satisfaction,
         JobInvolvement=job_involvement,
         JobLevel=job_level,
         JobSatisfaction=job_satisfaction,
         MonthlyIncome=monthly_income,
         OverTime=overtime,
         WorkLifeBalance=work_life_balance)
```
